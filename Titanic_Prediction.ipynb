```python
import numpy as np
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Read the data
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")

test_data.isna().sum()
```

```python
# Preprocess Fare column
test_data['Fare'] = test_data['Fare'].fillna(0)
```

```python
import math

def replace_cabin(cabin):
    if(isinstance(cabin, str)):
        return cabin[0]
```

```python
from sklearn.preprocessing import LabelEncoder

train_data_copy = train_data.copy()
test_data_copy = test_data.copy()

# Preprocessing the cabin column
train_data_copy['Deck'] = train_data['Cabin'].apply(replace_cabin)
test_data_copy['Deck'] = test_data['Cabin'].apply(replace_cabin)

# delete the Cabin feature
train_data = train_data_copy.drop('Cabin', axis = 1)
test_data = test_data_copy.drop('Cabin', axis = 1)

# Converting to numerical values
le = LabelEncoder()
le.fit(train_data_copy['Deck'])

encoding_list = []
encoding_list.append(train_data_copy['Deck'].unique())
encoding_list.append(list(le.transform(train_data_copy['Deck'].unique())))
train_data_copy['Deck'] = le.transform(train_data_copy['Deck'])
test_data_copy['Deck'] = le.transform(test_data_copy['Deck'])

train_data_copy['Deck'] = train_data_copy['Deck'].fillna((train_data_copy['Deck'].mean()))
test_data_copy['Deck'] = test_data_copy['Deck'].fillna((test_data_copy['Deck'].mean()))

train_data['Deck'] = train_data_copy['Deck']
test_data['Deck'] = test_data_copy['Deck']
```

```python
# Preprocess embarked column
# Fill embarked NaN with most frequent value
train_data.Embarked = train_data.Embarked.fillna('S')
test_data.Embarked = test_data.Embarked.fillna('S')

# Converting to numerical values
le = LabelEncoder()
le.fit(train_data['Embarked'])

encoding_list = []
encoding_list.append(train_data['Embarked'].unique())
encoding_list.append(list(le.transform(train_data['Embarked'].unique())))
train_data['Embarked'] = le.transform(train_data['Embarked'])
test_data['Embarked'] = le.transform(test_data['Embarked'])
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Deck", "Embarked", "Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators = 7, max_depth = 8, random_state = 1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
"Your submission was successfully saved!"
```
